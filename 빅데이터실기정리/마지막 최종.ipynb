## 제 1유형

```python
# airquality 데이터에 대해서 결측치가 가장 많은 변수를 찾아서 해당 결측치를 0으로 대치하고, 결측치를 제외한 평균과 0으로 대치한 후 평균의 차이를 구하시오.

import pandas as pd
import numpy as np

air = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv")
air.isnull().sum() # 가장 많은 건 Ozone 칼럼

before = air["Ozone"].sum()/air['Ozone'].count()

air["Ozone"] = air["Ozone"].fillna(0)
after = air["Ozone"].sum()/air['Ozone'].count()

# 평균의 차이
mean = abs(after - before)
print(mean)
```

```python
# Wind 변수에 대해서 Min-Max 정규화를 수행한 후의 평균값과 Z정규화흫 수행한 후 평균값의 차이를 구하시오.
air['minmax'] = (air['Wind'] - air['Wind'].min())/(air["Wind"].max()-air["Wind"].min())
air['Z-score'] = (air['Wind'] - air['Wind'].mean())/air["Wind"].std()

# <다른방법>
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
scaler = MinMaxScaler()
air["sk_MM"] = scaler.fit_transform(air[["Wind"]]) # fit_transform은 2차원 배열을 요구.
air['Sk_Z'] = Scaler.fit_transform(air[["Wind"]])

mean = abs(air['minmax'].mean()-air['Z-score'].mean())
sk_mean = abs(air['sk_MM'].mean() - air['Sk_Z'].mean())
print(mean, sk_mean)
```

```python
# 측정한 온도의 평균값이 77과 일치하는지 t-검정을 이용하여 확인하시오.

# 표본평균 X를 소수점 둘째자리까지 반올림하여 구하시오.
air = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv")
sample_mean = round(air['Temp'].mean(),2) # 77.88

# 가설검정을 위한 검정통계량을 소수점 둘째자리까지 반올림하여 구하시오.
from scipy.stats import ttest_1samp
t_statistic,p_value = ttest_1samp(air["Temp"],popmean = 77,alternative='two-sided')
t_statistic = round(t_statistic,2) # 1.15

# 통계량에 대한 p_value 값을 소수점 넷째자리까지 구하고, 유의수준 0.05내에서 결과를 논하시오.
if p_value > 0.05 :
  print(f'p_value:{round(p_value,4)}이고 0.05보다 크므로 귀무가설을 채택한다. 고로 평균값은 77과 같을 가능성이 있다.')
else:
  print(f'p_value:{round(p_value,4)}이고 0.05보다 작으므로 귀무가설을 기각한다. 고로 평균값은 77과 다를 가능성이 있다.')
```

## 제 2유형

```python
# 데이터 셋을 이용하여 파킨슨 병을 예측하는 로지스틱 회귀모형을 적용하여 생성하고, 이떄 파킨슨 병을 예측하는데 영향을 미치는 변수 중요한 순서대로 3개 선정하시오.
parkinson = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/Parkinsons.csv")

# 필요없는 칼럼 'name'제거
parkinson.drop(columns=['name'],inplace=True)

# Status는 카테고리 타입으로 변환
parkinson['status'] = parkinson['status'].astype('category')

# 트레이닝 셋과 테스트 셋의 비율은 9:1
from sklearn.model_selection import train_test_split
X = parkinson.drop(columns=['status'])
y = parkinson['status']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)

# min-max 정규화, 인덱스 생성 필수 : 정규화 하고 데이터프레임으로 바꾸면서 흐트러짐
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns, index=X_train.index)
X_test = pd.DataFrame(scaler.transform(X_test),columns=X.columns, index = X_test.index)

# 로지스틱 회귀분석을 위해 상수항을 추가
import statsmodels.api as sm
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

# 모델은 로지스틱 회귀분석 사용
model = sm.Logit(y_train,X_train)
result = model.fit(method='bfgs',maxiter=1000)
print(result.summary()) # 영향을 미치는 변수 3개 : P-value가 작은거,

# 임계치 설정 및 f1-score 확인
from sklearn.metrics import f1_score
def cutoff(y,threshold):
  Y = y.copy()
  Y[Y > threshold] = 1
  Y[Y <= threshold] = 0
  return Y.astype('int')

result = result.predict(X_test)
pred_0 = cutoff(result,0.5)
pred_1 = cutoff(result,0.8)

print(f1_score(y_test,pred_0))
print(f1_score(y_test,pred_1))
```
