```python
def hp_to_numeric(hp):
    if '200 - 300 HP' in hp:
        return 250
    elif '300 - 400 HP' in hp:
        return 350
    elif '400 - 500 HP' in hp:
        return 450
    elif '150 - 200 HP' in hp:
        return 175
    elif '500 - 600 HP' in hp:
        return 550
    elif '600 - 700 HP' in hp:
        return 650
    elif '700 - 800 HP' in hp:
        return 750
    elif '800 - 900 HP' in hp:
        return 850
    elif '900+ HP' in hp:
        return 950
    elif 'Less than 150 HP' in hp:
        return 100
    else:
        return None  # Unknown은 NaN

df['horsepower_numeric'] = df['horsepower'].apply(hp_to_numeric)
```

```python
df['no_of_cylinders'] = pd.to_numeric(df['no_of_cylinders'], errors='coerce')

count_df = df.groupby(['body_type', 'no_of_cylinders']).size().reset_index(name='count')

idx = count_df.groupby('body_type')['count'].idxmax()
result = count_df.loc[idx].reset_index(drop=True)

print(result)
```

```python
import pandas as pd

# 1. One-Hot Encoding (범주형 → 더미변수)
dummies = pd.get_dummies(df['regional_specs'], prefix='regional')

# 2. 수치형 변수와 합치기
df_encoded = pd.concat([df[['price_in_aed']], dummies], axis=1)

# 3. 상관계수 계산
corr_matrix = df_encoded.corr(method='pearson')

# 4. predicted_price와 각 더미변수 간 상관계수 출력
print(corr_matrix['price_in_aed'].drop('price_in_aed'))
```
