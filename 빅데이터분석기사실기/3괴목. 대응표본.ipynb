# 대응표본의 평균 검정(T-test)

- 두 관련된 데이터 집단 간의 평균 차이를 비교하고, 이 차이가 통계적으로 유의미한 지 확인하는 데 유용하다.

```python
from scipy import stats

data_before = [72,75,68,71,73]
data_after = [76,79,74,78,80]

t_statistic,p_value = stats.ttest_rel(data_before,data_after)

print(f'대응표본 t-test 결과:통계량 - {t_statistic}  //  p_value - {p_value}')

# 통계량이 음수인 경우 첫번째 그룹이 더 평균이 작음을 나타낸다.
# 통계값의 절댓값이 클수록 차이가 많이 남을 나타낸다.
```

# 실습 따라하기

제약회사에서 새로운 수면유도제를 개발하여 임상 실험을 해본 결과 다음과 같은 결과를 얻었다.

수면유도제 사용전 수면시간 [7,3,4,5,2,1,6,6,5,4]

수면유도제 사용후 수면시간 [8,4,5,6,2,3,6,8,6,5]

(1) 두 데이터 셋의 정규성 검정을 하시오.

(2) (1)의 결과를 기반으로  적절한 검증방법을 선태하여 검정을 수행하시오.

```python
# 정규성 검점

from scipy import stats

data_before = [7,3,4,5,2,1,6,6,5,4]
data_after = [8,4,5,6,2,3,6,8,6,5]

statistic_b,p_value_b = stats.shapiro(data_before)
statistic_a,p_value_a = stats.shapiro(data_after)

print(p_value_b,p_value_a)
# 결과 값이 0.05보다 크므로 귀무가설을 기각하지 못해 정규성이 있을 가능성이 있다고 판단된다.
```

```python
# 대응표본 t-test

t_statistic,p_value = stats.ttest_rel(data_before,data_after)

print(f'통계량  {t_statistic}  p-값  {p_value}')

if p_value > 0.05:
  print("귀무가설을 기각하지 못합니다. 전/후 차이가 없을 가능성이 있습니다.")
else:
  print("귀무가설을 기각합니다. 전/후 차이가 있을 가능성이 있습니다.")

# 통계량이 음수를 가지고 있어 전 데이터가 후 데이터보다 평균이 적다(수면시간이 적다).
```
