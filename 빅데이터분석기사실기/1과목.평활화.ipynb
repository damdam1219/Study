# 평활화
데이터의 분포를 매끄럽게 함으로써 데이터에서 중요하지 않은 것(잡음 등)을 제거하고 그 패턴을 알아내는 방법

- 목적

값의 변화를 매끄럽게 하여 데이터의 추세를 알기쉽게 하는 것이다.

---

**단순이동평균** 은 일정한 기간 동안의 값의 평균을 계산해서 시간에 따라 이동시키는 방식

예를 들어, 3일 단순이동평균은 다음과 같이 계산된다.
-> x1+x2+x3/3

즉, 현재 시점과 그 이전 2개의 값을 더해서 3으로 나눈 평균
다음 시점으로 갈 때는 가장 오래된 값을 제외하고 새 값을 포함시켜서 평균을 계산합니다. 그래서 "이동" 평균이라 부르고 평균값을 쓰기 떄문에 변동성을 줄일수 있다.

```python
# 데이터 준비: 1821년~ 1934년 까지의 캐나다 시라소니의 수
import pandas as pd

df= pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/lynx.csv")

df
```

```python
# 10년 단순이동령균(SMA)
import matplotlib.pyplot as plt
df["sma"]=df['value'].rolling(10).mean()

plt.plot(df['value'])
plt.plot(df["sma"])
plt.show()
```

**지수가중이동평균 EMA(Exponential Moving Average)**는 단순이동평균(SMA)의 단점을 보완한 기법입니다.

최근 데이터일수록 더 큰 가중치를 주고,

과거 데이터일수록 점점 작아지는 가중치를 적용해 평균을 계산합니다.

이로 인해 최근 변화에 더 민감하게 반응합니다.



```python
# 지수가중이동평균(EWM)
df["ewm"] = df['value'].ewm(10).mean()

plt.plot(df['value'])
plt.plot(df['ewm'])
plt.show()
```
