## 제 2유형

used_car 데이터 셋은 중고차와 관련한 정보를 가지고 있다. 주어진 학습용 데이터를 이용하여 중고차의 판매 가격을 예측하고, 이를 평가용 데이터에 적용하여 얻은 예측 결과를 다음과 같은 형식의 CSV 파일로 생성하시오.

- ID, Price

   1,12500
   
   2.62546

```python
car_train_y = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/used_car_y_train.csv")
car_train_X = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/used_car_X_train.csv")

car_train_X
```

```python
df = pd.merge(car_train_X,car_train_y,on='id')
df
```

```python
# 데이터 전처리
from sklearn.preprocessing import LabelEncoder

scaler = LabelEncoder()
list_car = ['transmission','fuelType']
for i in list_car:
  df[i] = scaler.fit_transform(df[i])

df
```

```python
df.drop(columns=['model']).corr()
```

```python
df.drop(columns=['model','id'],inplace=True)
```

```python
# 데이터 나누기

from sklearn.model_selection import train_test_split

X = df.drop(columns = ['price'])
y = df['price']

X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
# model made

import xgboost as xgb
model = xgb.XGBRegressor()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
print(np.sqrt(mse))  # 낮으면 낮을수록 모델의 예측이 더 정확하다.
```

```python
car_test_x = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/used_car_X_test.csv",encoding = 'cp949')

# 전처리

scaler = LabelEncoder()
list_car = ['transmission','fuelType']
for i in list_car:
  car_test_x[i] = scaler.fit_transform(car_test_x[i])

X_test = car_test_x.drop(columns=['model','id'])

pred_test = model.predict(X_test)
car_test_x
```

```python
df = pd.DataFrame(car_test_x['id'],columns= ['id'])
df['price'] = pred_test
df.to_csv('price_regressor.csv',index=False)
```
