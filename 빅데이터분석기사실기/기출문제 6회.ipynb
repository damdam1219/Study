````python
train = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_2_train.csv")
train
````

````python
from sklearn.preprocessing import LabelEncoder

scaler = LabelEncoder()
train['거주지역'] = scaler.fit_transform(train['거주지역'])
train['키'] = scaler.fit_transform(train['키'])
train['체격'] = scaler.fit_transform(train['체격'])
````

````python
# 감기약의 부작용 비율과 항람제 부작용 관찰값이 통계적으로 유의미하게 차이가 있는지 카이제곱 검정을 이용하여 검정통계량을 구하시오.

df = pd.DataFrame({
    '약제': ['항암제', '항암제', '항암제', '항암제', '감기약', '감기약', '감기약', '감기약'],
    '부작용': ['무증상', '속쓰림', '아픔', '조금아픔', '무증상', '속쓰림', '아픔', '조금아픔'],
    '비율': [0.55, 0.1, 0.15, 0.2, 0.7, 0.15, 0.05, 0.1]
})

table = pd.crosstab(index=df['부작용'], columns=df['약제'], values=df['비율'], aggfunc='sum')
````

````python
from scipy.stats import chi2_contingency

x_statistic,p_value,df,excepted = chi2_contingency(table)
````

````python
import pandas as pd
from scipy.stats import chi2_contingency

data = pd.DataFrame({
    '부작용': ['무증상', '속쓰림', '아픔', '조금아픔'],
    '항암제': [0.55, 0.1, 0.15, 0.2],
    '감기약': [0.7, 0.15, 0.05, 0.1]
})

# 카이제곱 검정용 교차표
table = data.set_index('부작용')

chi2_stat, p, dof, expected = chi2_contingency(table)

print("카이제곱 통계량:", chi2_stat)
print("p-value:", p)
````

주어진 데이터를 이용하여 다중선형회귀문제를 해결하시오.

````python
data = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_3_2.csv")
data
````

````python
# 선형회귀모델을 만들어 독립변수 Ozone의 회귀계수를 구하시오.

from statsmodels.formula.api import ols

formula = "Temp ~ Solar + Wind + Ozone"
model = ols(formula, data = data).fit()
coef = model.params["Ozone"]
print(coef)
````

````python
model.params
````

````python
model.summary()
````

````python
# Solar와 Ozone은 고정이고, Wind가 증가함에 따라 Temp가 감소하는지 검증하기위한 다중선형회귀분석을 수행하여 Wind의 회귀계수에 대한 p-value값을 구하시오.
# 위의 summary를 보면 알수 있듯이 Temp의 p값은 0.169이다.
````

````python
# Solar가 150, Wind 7.5, Ozone이 45일 경우, 예측갑소가 그에 대한 신뢰구간 95% 를 구하시오.

data2 = pd.DataFrame({'Ozone':[45],'Solar':[150],'Wind':[7.5]})
pred = model.get_prediction(data2)
summary = pred.summary_frame(alpha=0.05)
print(summary)
````
