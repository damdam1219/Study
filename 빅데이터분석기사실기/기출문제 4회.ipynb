## 제 2유형

Customer_Segmentation 데이터 셋은 고객 세분화 정보를 담고있다.주어진 훈련셋을 활용하여 고객이 속한 세그멘트를 예측하고 해당 예측결과를 다음과 같은 형식의 CSV파일로 저장하시오.

```python
import pandas as pd
customer_data_y = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/CS_Seg_y_train.csv")

from sklearn.preprocessing import LabelEncoder
scaler = LabelEncoder()
customer_data_y['Segmentation'] = scaler.fit_transform(customer_data_y['Segmentation'])

customer_data_y.drop(columns=['ID'],inplace=True)
```

```python
customer_data_x = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/CS_Seg_X_train.csv")

# customer_data_x['Ever_Married'] = scaler.fit_transform(customer_data_x['Ever_Married'])
# customer_data_x['Spending_Score'] = scaler.fit_transform(customer_data_x['Spending_Score'])
# customer_data_x['Graduated'] = scaler.fit_transform(customer_data_x['Graduated'])
# customer_data_x['Gender'] = scaler.fit_transform(customer_data_x['Gender'])
# customer_data_x = pd.get_dummies(customer_data_x,columns=['Profession'],dtype=int)
customer_data_x.drop(columns=['ID'],inplace=True)
customer_data_x.info()
for col in customer_data_x.select_dtypes('object').columns:
    customer_data_x[col] = customer_data_x[col].astype('category') # xgbooet는 category만 범주형으로 입력해야한다.
customer_data_x.info()
```

**상황에 따른 인코딩⭐**

| 모델 종류                                                 | 라벨 인코딩 적합 여부           | 이유                          |
| ----------------------------------------------------- | ---------------------- | --------------------------- |
| 트리 계열 (예: DecisionTree, RandomForest, XGBoost)        | **적합함**                | 숫자의 크기보다는 구분만 중요함           |
| 선형 모델, 회귀모델 (예: LinearRegression, LogisticRegression) | **주의 필요**              | 숫자 간 **서열 관계**를 잘못 학습할 수 있음 |
| 딥러닝 (예: MLP)                                          | 보통 One-Hot Encoding 추천 | 숫자 크기 의미 없음, 분산 고려 필요       |


```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(customer_data_x, customer_data_y, test_size=0.2, random_state=42)


# XGBoost 분류기 모델 생성 및 학습
model = xgb.XGBClassifier(enable_categorical=True, num_class = 4, eval_metric='mlogloss')  # 다중 분류
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print(f"정확도: {accuracy:.2f}")
```
